```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.auto import tqdm

from scipy.special import gamma, psi
def Pi(z): return gamma(z+1)
def diPi(z): return psi(z+1)
```

### [Gauss's Pi function](https://en.wikipedia.org/wiki/Gamma_function#Pi_function)

General definition  
The Gauss's Pi function is defined via a convergent improper integral:
$$ \Pi(z) = \int_{0}^{\infty} e^{-t} t^{z} dt$$
for $z \in \mathbb C, \qquad {\frak R}(z) > 0$
<br>
<br>
<br>
Reflection formula
$$ \Pi(z) \Pi(-z) = \frac{\pi z}{\sin{(\pi z)}} $$
<br>
<br>
Euler's definition as an infinite product
$$\Pi(z)=\prod_{n=1}^\infty \frac{\left(1+\frac{1}{n}\right)^z}{1+\frac{z}{n}}$$
  
Weierstrass's definition as an infinite product
$$\Pi(z) = e^{-\gamma z} \prod_{n=1}^\infty \left(1 + \frac{z}{n}\right)^{-1} e^{z/n}$$

[Relation to Gamma function](https://mathoverflow.net/questions/20960/why-is-the-gamma-function-shifted-from-the-factorial-by-1)   
$\Gamma(z) = \Pi(z-1) = \Pi(z) / z$  
$\Pi(z) = \Gamma(z+1) = \Gamma(z)*z$

```python
def remove_connecting(x,y):
  signchange = lambda x: np.where(x[:-1] * x[1:] < 0 )[0] +1
  pos = signchange(y)
  x[pos], y[pos] = np.nan, np.nan


fig, axs = plt.subplots(1,2, figsize=(15,5))

x1 = range(5)
y1 = [np.math.factorial(i) for i in x1]
x = np.linspace(-0.9,4,100)

plt.axes(axs[0])
plt.scatter(x1, y1, label='n!')
plt.plot(x, Pi(x), label='Π(x)', c='r')
plt.legend()
plt.title("interpolating factorial")



plt.axes(axs[1])
plt.scatter(x1, y1, s=20)
x = np.linspace(-6,4,20000)
y = Pi(x)
remove_connecting(x,y)
for i in range(-5, 0):  plt.axvline(x=i, color='b', ls='--')
plt.plot(x, y, c='r')

plt.ylim(-5,5)
plt.axhline(y=0, color='k', lw=0.7)
plt.axvline(x=0, color='k', lw=0.7)
plt.title("Gauss's Pi function")


plt.show()
```

### Family of Approximate Gamma functions

```python
lim = lambda f, x, eps=1e-4:  ( f(x-eps) + f(x+eps) ) / 2

def compare_gammafunc(axs, gamma_func1, offset1=0, gamma_func2=Pi, offset2=0):
  fname1, fname2 = gamma_func1.__name__, gamma_func2.__name__
  fname1, fname2 = map( lambda fname: 'Π' if fname=='Pi' else fname , [fname1,fname2])
  fname1 = fname1 + '(x' + ('' if offset1==0 else f'+{offset1}') + ')'
  fname2 = fname2 + '(x' + ('' if offset2==0 else f'+{offset2}') + ')'

  plt.axes(axs[0])

  x = np.linspace(-3,3,1000)
  plt.plot(x, gamma_func1(x+offset1), c='r')
  plt.axvline(x=0, color='k', lw=0.7)
  plt.title(fname1)


  plt.axes(axs[1])

  x1 = range(4)
  y1 = [np.math.factorial(i) for i in x1]
  x = np.linspace(-0.7,3,1000)

  plt.scatter(x1, y1, label='n!')
  plt.plot(x, gamma_func1(x+offset1), label=fname1, c='r')
  plt.plot(x, gamma_func2(x+offset2), label=fname2, c='y')
  plt.axvline(x=0, color='k', lw=0.7)
  plt.legend()
  plt.title(f'{fname1} , {fname2}')



  plt.axes(axs[2])
  x = np.arange(-5,6)
  ratio = lambda x: gamma_func1(x+offset1) / gamma_func2(x+offset2)
  plt.scatter(x, lim(ratio, x), s=20)
  x = np.linspace(-5,5,1000)
  plt.plot(x, ratio(x), c='r')
  plt.axvline(x=0, color='k', lw=0.7)
  plt.title(f"{fname1} / {fname2}")
```

#### [Hadamard's Gamma function](https://en.wikipedia.org/wiki/Hadamard%27s_gamma_function)

It is defined as:
$$ H(z) = \frac{1}{\Gamma \left(1-z \right)}
\left \{ \frac{d}{dz} \ln
\left ( \frac
{\Gamma \left ( \frac{1}{2} - \frac{z}{2} \right )}
{\Gamma \left ( 1 - \frac{z}{2} \right )}
\right ) 
\right \} $$
<br> <br>
Hadamard's gamma can also be expressed as
$$H(z) = \frac{\psi(1-\frac{z}{2})-\psi(\frac{1}{2}-\frac{z}{2})} {2\Gamma(1-z)}$$
<br> <br>
and as
$$  H(z) = \Gamma(z) \left[1 + \frac{\sin{\left(\pi z\right)}}{2\pi}
\left \{ \psi \left( \frac{z}{2} \right) 
-\psi \left( \frac{z+1}{2} \right) \right \}
\right]  $$
<br> <br>
Unlike the classical gamma function, Hadamard's gamma function H(z) is an entire function, i.e. it has no poles in its domain. It satisfies the functional equation
$$H(z+1) = zH(z) + \frac{1}{\Gamma(1-z)}$$

```python
def H(z):
  g = lambda z: (psi(z/2) - psi((z+1)/2))
  Q = lambda z: 1 + g(z) * np.sin(np.pi*z)/(2*np.pi)
  return gamma(z)*Q(z)

def H(z): return (psi(1-z/2)-psi(0.5-z/2))/(2*gamma(1-z))


fig, axs = plt.subplots(1,3, figsize=(25,5))

compare_gammafunc(axs, H, offset1=1)
plt.show()
```

#### [Luschny's Gamma function](https://www.luschny.de/math/factorial/hadamard/HadamardsGammaFunction.html)

$$  L(z) = \Pi(z) \left[1 - \frac{\sin{\left(\pi z\right)}}{\pi z}
\left \{ \frac{z}{2}  \left (\psi \left( \frac{z}{2} + \frac{1}{2} \right) 
- \psi \left( \frac{z}{2} \right) \right ) 
- \frac{1}{2} \right \}
\right]  $$
<br><br>
The identity analog to the definition of the Hadamard Gamma function is
$$  L(z) = \frac{1}{\Pi(-z)}
\left \{ \frac{1}{2} + 
z \frac{d}{dz} \ln 
\left ( \frac
{\Pi\left ( - \frac{z}{2} -\frac{1}{2} \right )}
{\Pi\left ( - \frac{z}{2} \right )}
\right ) 
\right \}  $$
<br><br>
Using the digamma function it can written as
$$ L(z) = \frac{1}{\Pi(-z)}
\left [ \frac{1}{2} + 
\frac{z}{2}
\left \{ 
\psi \left ( 1 - \frac{z}{2} \right ) -
\psi \left ( \frac{1}{2} - \frac{z}{2} \right ) 
\right \}
\right ] $$
<br><br>
Thus the functional equation of the Luschny's gamma function is
$$  L(z) = zL(z) + \frac{1}{2\Pi(z-1)}  $$

Luschny's second Gamma function

$$ L^{*}(z) = \Pi(z) \left ( \frac{1}{2} + \frac{1}{2z} \sqrt{z^2 + \frac{ (1 - e^{2z}) \sin(\pi z)}{e^{2z+1}\pi}} \right) $$
<br><br>
Reflection product
$$ L^{*}(z)L^{*}(-z) = \frac{\tanh(z)}{4z} $$
<br><br>
Super reflection product
$$ \frac{1}{L^{*}(z)L^{*}(-z)\Pi(z)\Pi(-z)} = \frac{4\sin(\pi z)}{\pi \tanh(z)} $$

```python
def L(z):
  g = lambda z: (z/2)*(psi(z/2 + 0.5) - psi(z/2)) - 0.5
  Q = lambda z: 1 - g(z) * np.sin(np.pi*z) / (np.pi*z)
  return Pi(z)*Q(z)

def L(z): return ( 0.5 + (z/2) * ( psi(1-(z/2)) - psi(0.5-(z/2)) ) ) / Pi(-z)

def L_2(z):
  Q = lambda z: 0.5 + (1/(2*z)) * np.sqrt(z**2 + (1 -np.exp(2*z)) * np.sin(np.pi*z) / ((np.exp(2*z) + 1)*np.pi))
  return Pi(z)*Q(z)

fig, axs = plt.subplots(1,3, figsize=(25,5))


compare_gammafunc(axs, L)
plt.show()
```

comparing L(x) with L*(x)

```python
x = np.linspace(-6,6,1000)
plt.plot(x, (L_2(x)-L(x)) / Pi(x))
plt.axhline(y=0, c='k', lw=0.5)
plt.title('(L*(x)- L(x)) / Π(x)')
plt.show()

x = np.linspace(0,5,1000)
err1 = abs(Pi(x)-L(x))
err1 = err1[~np.isnan(err1)].cumsum()
err2 = abs(Pi(x)-L_2(x))
err2 = err2[~np.isnan(err2)].cumsum()
print(f'err of L(z) {err1[-1]:0.3f} from 0 to 5')
print(f'err of L*(z) {err2[-1]:0.3f} from 0 to 5')
# plt.plot(err1, label='abs(Π(x)-L(x))')
# plt.plot(err2, label='abs(Π(x)-L*(x))')
# plt.legend()
# L(x) is more accurate than L(z)
```

Luschny's factorial function L(x) approximates the non integral values x > 0 of the factorial function better then Hadamard's Gamma function H(x) the non integral values x > 0 of the Gamma function

```python
x = np.linspace(0,6)

plt.plot(x, gamma(x)-H(x), label='Γ(x)-H(x)')
plt.plot(x, Pi(x)-L(x), label='Π(x)-L(x)')
plt.axhline(y=0, color='k', lw=0.5)

plt.legend()
plt.show()
```

#### [Simpler Gamma function](https://arxiv.org/abs/2107.11330)

Simple gamma function using a single digamma

$$ K(z) = \Gamma(z)\left ( 1 + \frac{\sin(2 \pi z)}{2 \pi z} \psi(z) \right ) $$
<br><br>
The recurrence relation satisfied by $K(z)$ is the same as that of $\Gamma(z)$ plus an extra term which vanishes at all integers and half-integers
$$ K(z+1) = zK(z) + \Gamma(z)\frac{\sin(2 \pi z)}{2 \pi} $$
<br><br>
This can also be written as
$$ K(z+1) = zK(z) + \frac{\cos(\pi z)}{\Gamma(1-z)} $$

```python
def K(z):
  Q = lambda z: 1 + psi(z) * np.sin(2*np.pi*z) / (2*np.pi)
  return gamma(z)*Q(z)

def K(z): return gamma(z)*(1 + psi(z) * np.sin(2*np.pi*z) / (2*np.pi))

fig, axs = plt.subplots(1,3, figsize=(25,5))

compare_gammafunc(axs, K, offset1=1)
plt.show()
```

An example of the phenomenon that K(z) and the Euler gamma are equal wherever one of them has an extremum

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))


plt.axes(axs[0])

x = np.linspace(-6,4,100)
plt.plot(x, K(x+1), c='k', label='K(x+1)')

x = np.linspace(-6,4,20000)
y = Pi(x)
remove_connecting(x,y)
plt.plot(x, y, c='b', ls='--', label='Π(x)')
for i in range(-5,0): plt.plot(x, Pi(x), c='b', ls='--')

plt.ylim(-5,5)
plt.axhline(y=0, color='k', lw=0.7)
plt.axvline(x=0, color='k', lw=0.7)
plt.legend()



plt.axes(axs[1])

x = np.linspace(-2.65,-2.45,100)
plt.plot(x, K(x+1), c='k', label='K(x+1)')
plt.plot(x, Pi(x), c='b', ls='--', label='Π(x)')
plt.legend()


plt.show()
```

### [Reciprocal Pi function](https://en.wikipedia.org/wiki/Reciprocal_gamma_function)

Since the Pi function is meromorphic and nonzero everywhere in the complex plane, its reciprocal is an entire function

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))

plt.axes(axs[0])
x = np.linspace(-5.1,5,1000)

plt.plot(x, 1/Pi(x))
plt.axhline(y=0, color='k', lw=0.7)
plt.axvline(x=0, color='k', lw=0.7)
plt.title('y = 1/Π(x)')


plt.axes(axs[1])
x = np.linspace(-5.1,5,1000)
plt.plot(x, 1/Pi(x), label='1/Π(x)')

x = np.linspace(-5,4,10000)
y =Pi(x)
remove_connecting(x,y)
plt.plot(x, y, label='Π(x)')
for i in range(-4,0): plt.axvline(x=i, color='b', ls='--')

plt.ylim(-5,5)
plt.axhline(y=0, color='k', lw=0.7)
plt.axvline(x=0, color='k', lw=0.7)
plt.legend()


plt.show()
```

```python
from matplotlib.colors import hsv_to_rgb, rgb_to_hsv
```

```python
# define size
# using top left and bottom right coordinates
img_size = ((0,0),(512,512))
plot_size = ((-6,-6),(6,6))
(plot_x0,plot_y0), (plot_x1,plot_y1) = plot_size
(img_x0,img_y0), (img_x1,img_y1) = img_size

# define parameters
plot_xStep = (plot_x1-plot_x0)/(img_x1-img_x0)
plot_yStep = (plot_y1-plot_y0)/(img_y1-img_y0)
params = {"plot_xStep":plot_xStep, "plot_yStep":plot_yStep}
```

```python
def plot_func(plot_size,img_size,params, gamma_func=Pi, color_strength=True):

  (plot_x0,plot_y0), (plot_x1,plot_y1) = plot_size
  (img_x0,img_y0), (img_x1,img_y1) = img_size
  img_w = img_x1 - img_x0
  img_h = img_y1 - img_y0

  img = np.zeros((img_w, img_h, 3))

  x = np.arange(plot_x0, plot_x1+params["plot_xStep"], params["plot_xStep"])
  y = np.arange(plot_y0, plot_y1+params["plot_xStep"], params["plot_yStep"])
  xx, yy = np.meshgrid(x,y)
  zz = xx + yy*1j
  zz = 1/gamma_func(zz)
  zz[np.isnan(zz)] = 0

  # color depends on the angle of the roots
  angle = ( (0 + np.angle(zz, deg=True)) %360)/360   # [0,1]
  h = angle

  if color_strength:
    s = np.log(1 + abs(zz))
    s = 1 - s / (4 + s)   # 1 for small values, 0 for large values

    v = abs(zz)
    v = v / (0.1 + v)  # 0 for small values, 1 for large values
  else:
    s =  np.ones_like(zz, dtype=float)
    v =  np.ones_like(zz, dtype=float)

  img = np.dstack((h,s,v))  
  return hsv_to_rgb(img)
```

```python
img = plot_func(plot_size,img_size,params)
# plt.imshow(img)   # uncomment to see raw img
```

```python
dpi = plt.rcParams['figure.dpi'] # get the default dpi value
dpi=50
fig_size = np.array(img.shape[:2])//dpi # figure size

fig, ax = plt.subplots(figsize=fig_size)
ax.imshow(img)

def set_axes(ax):
  # xlen = 2*len(ax.get_xticks())
  # ylen = 2*len(ax.get_yticks())
  xlen, ylen = 12, 12

  xlen += (xlen+1)%2; ylen += (ylen+1)%2
  xticks = np.linspace(img_x0,img_x1-1,xlen)
  yticks = np.linspace(img_y0,img_y1-1,ylen)
  xlabels = np.round(np.linspace(plot_x0,plot_x1,xlen),2)
  ylabels = np.round(np.linspace(plot_y1,plot_y0,ylen),2)
  ylabels = [str(l)+'i' for l in ylabels]

  plt.xticks(ticks=xticks, labels=xlabels, rotation=-90)
  plt.yticks(ticks=yticks, labels=ylabels)
  plt.xlabel("Re")
  plt.ylabel("Im")

set_axes(ax)
ax.set_title("Reciprocal Pi function")
ax.set_frame_on(False)
```

#### compare reciprocal gamma functions

```python
img_size = ((0,0),(512,512))
plot_size = ((-18,-18),(18,18))
(img_x0,img_y0), (img_x1,img_y1) = img_size
(plot_x0,plot_y0), (plot_x1,plot_y1) = plot_size

plot_xStep = (plot_x1-plot_x0)/(img_x1-img_x0)
plot_yStep = (plot_y1-plot_y0)/(img_y1-img_y0)
params = {"plot_xStep":plot_xStep, "plot_yStep":plot_yStep}

fig, axs = plt.subplots(3,2, figsize=(12,20))
fig.delaxes(axs[2,1])

plt.axes(axs[0,0])
img = plot_func(plot_size,img_size,params, color_strength=False)
set_axes(axs[0,0])
plt.title('y = 1/Π(x)')
plt.imshow(img)

plt.axes(axs[0,1])
img = plot_func(plot_size,img_size,params, lambda x: H(x+1), False)
set_axes(axs[0,1])
plt.title('y = 1/H(x+1)')
plt.imshow(img)

plt.axes(axs[1,0])
img = plot_func(plot_size,img_size,params, L, False)
set_axes(axs[1,0])
plt.title('y = 1/L(x)')
plt.imshow(img)

plt.axes(axs[1,1])
img = plot_func(plot_size,img_size,params, lambda x: K(x+1), False)
set_axes(axs[1,1])
plt.title('y = 1/K(x+1)')
plt.imshow(img)

plt.axes(axs[2,0])
img = plot_func(plot_size,img_size,params, lambda x: L_2(x), False)
set_axes(axs[2,0])
plt.title('y = 1/L*(x)')
plt.imshow(img)

plt.show()
```
